# ソート処理をする

```python
df3 = df1.copy()
df3 = df3.sort_values('貸出日時')

df3.head()
```

```python
df4 = df3.reset_index(drop=True)
df4.head()
```

# queryメソッド

```python
x1 = df3.query('顧客ID == 459')
x1.head()
```

```python
# 2行目の顧客IDを取得する
cust_id = df4.iloc[1]['顧客ID']
print(cust_id)
```

```python
# @マークで、列名ではなく、変数として認識させる
x3 = df3.query('顧客ID == @cust_id')
x3.head()
```
