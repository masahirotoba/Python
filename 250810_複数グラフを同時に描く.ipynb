```python
import matplotlib.pyplot as plt
iris = load_iris()
x = iris.data
y = iris.target
columns = iris.feature_names
print(x[:5])
```

# iris.targetとは
iris.target はアイリスの花の種類を数値で表したデータです。

150個のサンプル（花）
各サンプルに4つの特徴量（iris.data）
各サンプルに1つの正解ラベル（iris.target）

# 複数グラフを同時に描く

```python
plt.figure(figsize=(15, 5))

for i in range(1, 4):
  ax = plt.subplot(1, 3, i)
  ax.scatter(x[:, 0], x[:, i], c=y, cmap='rainbow')
  ax.set_title(columns[0] + 'vs' + columns[i])
  ax.grid()

plt.tight_layout()
plt.show()
```
