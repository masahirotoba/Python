```python
values = [1,2,3,4,5,6]
x_pos = np.arange(len(values))

plt.bar(x_pos, values)
plt.show()
```

x_pos = np.arange(len(values))

X軸の位置として0, 1, 2, 3, 4, 5を生成
len(values)は6なので、np.arange(6)で[0,1,2,3,4,5]の配列を作成
