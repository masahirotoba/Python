```python
import pandas as pd
from IPython.display import display
```

# 表示方法を設定する

```python
# 小数点以下4桁で固定
pd.options.display.float_format = '{:.4f}'.format

# データフレームで全ての項目を表示
pd.set_option('display.max_columns', None)
```
