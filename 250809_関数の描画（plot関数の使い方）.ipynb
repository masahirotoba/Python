# plt.plotってなんですか？
pltはmatplotlib.pyplotモジュール全体を指すオブジェクトになります。

```python
x = np.linspace(0, 2, 101)
y1 = np.sqrt(x)
y2 = x ** 2
```

```python
plt.plot(x, y1, label='sqrt(x)')
plt.plot(x, y2, label='x^2')

plt.legend()

plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.title('sqrt(x) and x^2')
plt.show()
```
