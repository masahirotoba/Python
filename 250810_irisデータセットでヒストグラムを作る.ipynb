# ライブラリをインポート

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import numpy as np
```

# Irisデータセットの読み込み

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
```

# 基本情報を出力

```python
print(f"データの形: {df.shape}")
print(f"カラム名: {list(df.columns)}")
```

# 基本統計量

```python
print(df.describe())
```

# 度数分布表を作る

```python
frequency_table = df['species'].value_counts()
print(frequency_table)
```

# 度数分布表を出力

```python
petal_length_bins = pd.cut(df['petal length (cm)'], bins=5)
petal_freq = petal_length_bins.value_counts().sort_index()
print(petal_freq)
```
