```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
pittsburgh_bridges = fetch_ucirepo(id=18)

# data (as pandas dataframes)
X = pittsburgh_bridges.data.features
y = pittsburgh_bridges.data.targets

# metadata
print(pittsburgh_bridges.metadata)

# variable information
print(pittsburgh_bridges.variables)
```

```python
df = pd.concat([X, y], axis=1)
df.head()
```

```python
df.hist()
plt.show()
```

```python
df.groupby('PURPOSE').size()
```

```python
# 
purpose_counts = df.groupby('PURPOSE').size()
purpose_counts.plot(kind='bar')
plt.title('COUNTS by PURPOSE')
plt.xlabel('PURPOSE')
plt.ylabel('COUNTS')
plt.xticks(rotation=45)  # x45
plt.tight_layout()  # 
plt.show()
```

```python
# 
purpose_counts = df.groupby('PURPOSE').size()
purpose_counts.plot(kind='bar')
plt.title('COUNTS by PURPOSE')
plt.xlabel('PURPOSE')
plt.ylabel('COUNTS')
plt.show()
```

```python
# 
purpose_counts = df.groupby('PURPOSE').size()
purpose_counts.plot(kind='pie')
plt.title('COUNTS by PURPOSE')
plt.xlabel('PURPOSE')
plt.ylabel('COUNTS')
plt.show()
```

```python
# 
purpose_counts = df.groupby('PURPOSE').size()
purpose_counts.plot(kind='barh')
plt.title('COUNTS by PURPOSE')
plt.xlabel('PURPOSE')
plt.ylabel('COUNTS')
plt.show()
```
