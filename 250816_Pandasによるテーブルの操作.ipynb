# ライブラリをインポート

```python
import pandas as pd
from IPython.display import display
```

# 表示方法を設定する

```python
# 小数点以下4桁で固定
pd.options.display.float_format = '{:.4f}'.format

# データフレームで全ての項目を表示
pd.set_option('display.max_columns', None)
```

# CSVをインポート（復習）

```python
# Googleドライブにアクセスするためのモジュールをインポートする
from google.colab import drive

df_csv = pd.read_csv('売上データ.csv')
display(df_csv)
```

```python
# 元に影響が出ないようにする
df2 = df_csv.copy()

# 原価列を削除する
df2 = df2.drop(columns = '原価')

display(df2)
```

```python
df2 = df2.drop(columns='売上')

display(df2)
```

```python
売上 = df2['数量'] * df2['単価']
display(売上)
```

```python
# 影響が出ないようにコピー
df3 = df2.copy()

# 売上列を追加する
df3['売上'] = 売上

display(df3)
```

```python
df4 = df2.copy()

# 売上列を作る
売上 = df4['単価']*df4['数量']

df4.insert(2, '売上', 売上)

display(df4)
```
