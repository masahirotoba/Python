```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'o')
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'o')
plt.axis([0,6,0,10])
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', label='y=x')
plt.plot(t, t**2, 'bs', label='y=x**2')
plt.plot(t, t**3, 'g^', label='y=x**3')
plt.legend()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.show()
```

```python
x =np.random.randint(1,100,50)
y =np.random.randint(1,100,50)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randint(1, 100, 50)
y = np.random.randint(1, 100, 50)
plt.scatter(x, y, c=x+y, cmap='viridis')  # x+y
plt.colorbar()  # 
plt.show()
```

```python
x = np.random.randint(1, 100, 50)
y = np.random.randint(1, 100, 50)

plt.scatter(range(50), x, c='red', label='x values')    # x
plt.scatter(range(50), y, c='blue', label='y values')   # y
plt.legend()  # 
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
values = [1,2,3,4,5,6]
x_pos = np.arange(len(values))

plt.bar(x_pos, values)
plt.show()
```

```python
values = [1,2,3,4,5,6]
x_pos = np.arange(len(values))
objects = ('A', 'B', 'C', 'D', 'E', 'F')

plt.bar(x_pos, values)
plt.xticks(x_pos, objects)
plt.show()
```
